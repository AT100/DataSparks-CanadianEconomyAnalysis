```python
import pandas as pd
import numpy as np
from os import listdir
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import preprocessing
import boto3
from io import StringIO
```

```python
s3_obj = boto3.client('s3')
s3_exp_obj = s3_obj.get_object(Bucket='mysparks', Key="OUTPUT-Folder/gdp_pred_input/part-00000-42f94478-ef14-4978-8d86-f6696c1b1bf3-c000.csv")
s3_exp_data = s3_exp_obj['Body'].read().decode('utf-8')
```

```python
data = pd.read_csv(StringIO(s3_exp_data))
data
```

```python
x = data.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data_scaled = pd.DataFrame(x_scaled)
#data_scaled.rename(index=str).index
data_scaled.columns = ['YEAR','Hours_Worked','labour_productivity','CPI','GDP_millions','Mercahtn_Trade_millions','household_expenditure_millions']
data_scaled
```

```python
#Manually splitting train and test data to get 2010 to 2017 for training and 2018 to 2020 for testing
train_data = data_scaled[data_scaled['YEAR']<0.8]
test_data = data_scaled[data_scaled['YEAR']>=0.8]
```

```python
X_train = train_data.drop(['YEAR','GDP_millions'],axis=1)
X_test = test_data.drop(['YEAR','GDP_millions'],axis=1)
```

```python
plt.matshow(X_train.corr())
plt.show()
```

```python
y_train = train_data['GDP_millions']
y_test = test_data['GDP_millions']
```

```python
X_train
```

```python
#Using Elastic Net CV since the values in feature set are strongly correlated


alphas = [0.0001, 0.001, 0.01, 0.1, 0.3, 0.5, 0.7, 1]
elastic_cv=ElasticNetCV(alphas=alphas, cv=5)
model = elastic_cv.fit(X_train, y_train)
#clf =GradientBoostingRegressor()
#model = clf.fit(X_train, y_train)
```

```python
ypred = model.predict(X_test)
score = model.score(X_test, y_test)
mse = mean_squared_error(y_test, ypred)
print("R2:{0:.4f}, MSE:{1:.4f}, RMSE:{2:.4f}"
      .format(score, mse, np.sqrt(mse)))
```

```python
x_ax = range(len(X_test))
plt.scatter(x_ax, y_test, s=5, color="blue", label="original")
plt.plot(x_ax, ypred, lw=0.8, color="red", label="predicted")
plt.legend()
plt.show()
```

### GDP values predicted for years 2018, 2019 and 2020 is almost in line with the actual values
